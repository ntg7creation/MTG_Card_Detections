```python
import cv2
from IPython.display import Image
import numpy as np
```

```python
img=cv2.imread('symbols/Mana.png')
print(img.shape) # Print image shape
cv2.imshow("original", img)
 
# Cropping an image
for i in range(3):
    for j in range(10):
        cropped_image = img[i*105:i*105+100,j*105:j*105+100]
        # Display cropped image
        #cv2.imshow("cropped", cropped_image)   
        # Save the cropped image
        cv2.imwrite("symbols/mana_"+str(i)+"_"+str(j)+".jpg", cropped_image)
        #Image(cropped_image) 
```
